```python
import numpy as np
import matplotlib.pyplot as plt
import Rocket as r
```

```python
help(r.Rocket)
```

```python
env=r.Rocket()

env.reset()


data=np.zeros((100,11))

t=np.arange(0,100)*0.001*50

data[:,0]=t
env.step(1)

for i in range(t.size):
    data[i,1:]=np.array(env.step(0)[:10])
```

```python
plt.figure(1,figsize=(10,6))

plt.title("Position x",fontsize=20)

plt.xlabel("t",fontsize=16)
plt.ylabel("x",fontsize=16)


plt.plot(t,data[:,1],label="Numerical integration")
#plt.plot(t,2.0/(4.0+t*t-2.0*np.cos(t)-2.0*t*np.sin(t)),label="Solution")

plt.legend(loc="upper left")
```

```python
plt.figure(1,figsize=(10,6))

plt.title("Position y",fontsize=20)

plt.xlabel("t",fontsize=16)
plt.ylabel("y",fontsize=16)


plt.plot(t,data[:,2])
```

```python
plt.figure(1,figsize=(10,6))

plt.title(r"Angle $\phi$",fontsize=20)

plt.xlabel("t",fontsize=16)
plt.ylabel(r"$\phi$",fontsize=16)

#plt.ylim(-np.pi,np.pi)


plt.plot(t,data[:,3])
```

```python
plt.figure(1,figsize=(10,6))

plt.title(r"Velocity $\dot{x}$",fontsize=20)

plt.xlabel("t",fontsize=16)
plt.ylabel(r"$\dot{x}$",fontsize=16)




plt.plot(t,data[:,4])
```
